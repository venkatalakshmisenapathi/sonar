IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

IMPORTING DATASET

```python
dataset = pd.read_csv('/content/sonar data.csv',header= None)
```

```python
dataset.head() #First 5 rows of data in the dataset
```

```python
dataset.tail()  #Last 5 rows of data in the dataset
```

```python
dataset.isnull().sum() #Checking the presence of null values in the dataset
```



*   R --> Rock
*   M --> Mine



```python
dataset[60].value_counts()
```

```python
X = dataset.drop(columns=60, axis=1)
Y = dataset[60]
```

Spliting the dataset into trainig set and testing set

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2,random_state=2,stratify=Y)
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
X_train_pred = model.predict(X_train)
train_accuracy = accuracy_score(X_train_pred,y_train)
print("The accuracy score on the training dataset is",round(train_accuracy*100,2))
```

```python
X_test_pred = model.predict(X_test)
test_accuracy = accuracy_score(X_test_pred,y_test)
print("The accuracy score on the testing dataset is",round(test_accuracy*100,2))
```

Model prediction with unseen data

```python
user_input = input("Enter the data:")
input_data = [float(x) for x in user_input.split(',')]
input_as_array = np.asarray(input_data)
input_reshape = input_as_array.reshape(1,-1)
prediction = model.predict(input_reshape)
if prediction[0] == 'R':
  print("It is Rock.")
else:
  print("It is Mine")
```
